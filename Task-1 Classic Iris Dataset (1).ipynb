## Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
df_train, df_test = train_test_split(df,test_size= 0.25)
```

```python

df_train.shape
```

```python
df_test.shape
```

```python
df_train.head()
```

## Preparing our dataset for modeling

```python
x_train = df_train.drop(columns=['target', 'target_names']).values
y_train = df_train['target'].values
```

```python
x_train.shape
```

```python
y_train
```

## Modeling - Simple manual model

```python
def single_feature_model(petal_length):
    if petal_length < 2.5:
        return 0
    elif petal_length < 5:
        return 1
    else:
        return 2
```

```python
manual_model = [single_feature_model(val) for val in x_train[:, 2]]
```

```python
manual_model_accuracy = np.mean(manual_model == y_train)
```

```python
print(f"Manual model accuracy is {manual_model_accuracy * 100:.2f}%")
```

## Modeling - Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

### Using a validation set to evaluate a model

```python
xt,xv,yt,yv = train_test_split(x_train,y_train,test_size=0.25)
```

```python
xt.shape
```

```python
xv.shape
```

```python
model = LogisticRegression()
```

```python
model.fit(xt,yt)
```

```python
y_pred = model.predict(xv)
```

```python
np.mean(y_pred == yv)
model.score(xv,yv)
```

 ### using cross validation to evaluate our models 

```python
from sklearn.model_selection import cross_val_predict,cross_val_score
```

```python
model = LogisticRegression(max_iter=200)
```

```python
accs = cross_val_score(model,x_train,y_train,cv=5)
```

```python
np.mean(accs)
```

# Final Model

```python
model = LogisticRegression(max_iter = 200, C=2)
```

```python
x_test = df_test.drop(columns=['target', 'target_names']).values
y_test = df_test['target'].values
```

```python
model.fit(x_train, y_train)
```

```python
y_test_pred = model.predict(x_test)
```

```python
test_accurate = y_test_pred == y_test
test_accuracy = np.mean(test_accurate)
```

```python
print(f"The Accuracy of the Machine Learning model trained on classic iris dataset is: {test_accuracy*100:.2f}%")
```
